```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
model_name = "microsoft/DialoGPT-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
prompt = "Hello! How are you?"

inputs = tokenizer.encode(prompt + tokenizer.eos_token, return_tensors="pt")
```

```python
output = model.generate(
    inputs,
    max_new_tokens=50,
    pad_token_id=tokenizer.eos_token_id,
    do_sample=True,
    temperature=0.7
)
```

```python
response = tokenizer.decode(output[:, inputs.shape[-1]:][0], skip_special_tokens=True)

print(response)
```

```python
from transformers import PegasusTokenizer, PegasusForConditionalGeneration
```

```python
model_name = "google/pegasus-xsum"

tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name)
```

```python
text = """
Artificial Intelligence is transforming industries across the world.
It is used in healthcare, finance, education, transportation, and manufacturing.
Machine learning and deep learning are important branches of AI that help computers learn from data.
"""
```

```python
inputs = tokenizer(
    text,
    truncation=True,
    padding="longest",
    return_tensors="pt"
)
```

```python
summary_ids = model.generate(
    inputs["input_ids"],
    max_new_tokens=30
)
```

```python
summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print(summary)
```
