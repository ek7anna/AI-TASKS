```python
import spacy

# Load the English model
nlp = spacy.load("en_core_web_sm")

# Input text
text = """
Barack Obama visited New York last week.
He met Satya Nadella at Microsoft headquarters.
"""

# Process the text
doc = nlp(text)

# Print named entities
print("Named Entities Found:\n")

for ent in doc.ents:
    print(f"Entity: {ent.text}")
    print(f"Label : {ent.label_}")
    print("-" * 30)
```

```python
from spacy import displacy

# Display entities with colors
displacy.render(doc, style="ent", jupyter=True)
```

```python
from transformers import pipeline

# Load the pre-trained BERT NER pipeline
ner = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    tokenizer="dslim/bert-base-NER",
    aggregation_strategy="simple"
)

# Input text
text = """
Barack Obama visited New York last week.
He met Satya Nadella at Microsoft headquarters.
"""

# Perform Named Entity Recognition
results = ner(text)

# Display results
print("Named Entities Found:\n")

for entity in results:
    print(f"Entity : {entity['word']}")
    print(f"Label  : {entity['entity_group']}")
    print(f"Score  : {entity['score']:.4f}")
    print("-" * 30)
```
